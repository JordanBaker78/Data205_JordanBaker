```python
import pandas as pd
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from numpy._core.fromnumeric import std
import scipy.stats as stats
```

```python
stats_2023 = pd.read_csv('https://raw.githubusercontent.com/JordanBaker78/Data205_JordanBaker/refs/heads/main/Project_Datasets/Stats-2023.csv')
stats_2022 = pd.read_csv('https://raw.githubusercontent.com/JordanBaker78/Data205_JordanBaker/refs/heads/main/Project_Datasets/Stats-2022.csv')
stats_2021 = pd.read_csv('https://raw.githubusercontent.com/JordanBaker78/Data205_JordanBaker/refs/heads/main/Project_Datasets/Stats-2021.csv')
stats_2020 = pd.read_csv('https://raw.githubusercontent.com/JordanBaker78/Data205_JordanBaker/refs/heads/main/Project_Datasets/Stats-2020.csv')
stats_2019 = pd.read_csv('https://raw.githubusercontent.com/JordanBaker78/Data205_JordanBaker/refs/heads/main/Project_Datasets/Stats-2019.csv')
```

```python
stats_df = pd.concat((stats_2023, stats_2022, stats_2021, stats_2020, stats_2019))
stats_df.tail()
```

```python
rush_avg = stats_df['rush_yds'] / stats_df['rush_att']
stats_df.insert(8, 'rush_avg', rush_avg)
stats_df.head()
```

```python
rush_td_per_att = stats_df['rush_tds'] / stats_df['rush_att']
fumbles_lost_per_att = stats_df['fumbles_lost'] / stats_df['rush_att']
rush_success = stats_df['rush_avg'] + (4 * rush_td_per_att) - (4 * fumbles_lost_per_att)
stats_df.insert(9, 'rush_success', rush_success)
stats_df.head()
```

```python
stats_2019 = stats_df[stats_df['season'] == 2019]
stats_2020 = stats_df[stats_df['season'] == 2020]
stats_2021 = stats_df[stats_df['season'] == 2021]
stats_2022 = stats_df[stats_df['season'] == 2022]
stats_2023 = stats_df[stats_df['season'] == 2023]
stats_2019.head()
```

```python
rushing_2019 = stats_2019[['name', 'rush_att', 'rush_yds', 'rush_avg', 'rush_tds', 'rush_success', 'fumbles', 'fumbles_lost']]
rushing_2020 = stats_2020[['name', 'rush_att', 'rush_yds', 'rush_avg', 'rush_tds', 'rush_success', 'fumbles', 'fumbles_lost']]
rushing_2021 = stats_2021[['name', 'rush_att', 'rush_yds', 'rush_avg','rush_tds', 'rush_success', 'fumbles', 'fumbles_lost']]
rushing_2022 = stats_2022[['name', 'rush_att', 'rush_yds', 'rush_avg', 'rush_tds', 'rush_success', 'fumbles', 'fumbles_lost']]
rushing_2023 = stats_2023[['name', 'rush_att', 'rush_yds', 'rush_avg', 'rush_tds', 'rush_success', 'fumbles', 'fumbles_lost']]
rushing_total = stats_df[['season', 'name', 'rush_att', 'rush_yds', 'rush_avg', 'rush_tds', 'rush_success', 'fumbles', 'fumbles_lost']]
rushing_total.head()
```

```python
all_teams_total_rush_2019 = stats_2019.groupby('name')['rush_yds'].sum().reset_index()
all_teams_total_rush_2020 = stats_2020.groupby('name')['rush_yds'].sum().reset_index()
all_teams_total_rush_2021 = stats_2021.groupby('name')['rush_yds'].sum().reset_index()
all_teams_total_rush_2022 = stats_2022.groupby('name')['rush_yds'].sum().reset_index()
all_teams_total_rush_2023 = stats_2023.groupby('name')['rush_yds'].sum().reset_index()
```

```python
all_teams_rush_success_2019 = stats_2019.groupby('name')['rush_success'].mean().reset_index()
all_teams_rush_success_2020 = stats_2020.groupby('name')['rush_success'].mean().reset_index()
all_teams_rush_success_2021 = stats_2021.groupby('name')['rush_success'].mean().reset_index()
all_teams_rush_success_2022 = stats_2022.groupby('name')['rush_success'].mean().reset_index()
all_teams_rush_success_2023 = stats_2023.groupby('name')['rush_success'].mean().reset_index()
```

```python
all_teams_rush_tds_2019 = stats_2019.groupby('name')['rush_tds'].sum().reset_index()
all_teams_rush_tds_2020 = stats_2020.groupby('name')['rush_tds'].sum().reset_index()
all_teams_rush_tds_2021 = stats_2021.groupby('name')['rush_tds'].sum().reset_index()
all_teams_rush_tds_2022 = stats_2022.groupby('name')['rush_tds'].sum().reset_index()
all_teams_rush_tds_2023 = stats_2023.groupby('name')['rush_tds'].sum().reset_index()
```

```python
all_teams_rush_avg_2019 = stats_2019.groupby('name')['rush_avg'].mean().reset_index()
all_teams_rush_avg_2020 = stats_2020.groupby('name')['rush_avg'].mean().reset_index()
all_teams_rush_avg_2021 = stats_2021.groupby('name')['rush_avg'].mean().reset_index()
all_teams_rush_avg_2022 = stats_2022.groupby('name')['rush_avg'].mean().reset_index()
all_teams_rush_avg_2023 = stats_2023.groupby('name')['rush_avg'].mean().reset_index()
```

```python
all_teams_rush_fumbles_lost_2019 = stats_2019.groupby('name')['fumbles_lost'].sum().reset_index()
all_teams_rush_fumbles_lost_2020 = stats_2020.groupby('name')['fumbles_lost'].sum().reset_index()
all_teams_rush_fumbles_lost_2021 = stats_2021.groupby('name')['fumbles_lost'].sum().reset_index()
all_teams_rush_fumbles_lost_2022 = stats_2022.groupby('name')['fumbles_lost'].sum().reset_index()
all_teams_rush_fumbles_lost_2023 = stats_2023.groupby('name')['fumbles_lost'].sum().reset_index()
```

```python
total_rush_with_rank_19 = all_teams_total_rush_2019.copy()
total_rush_with_rank_19['rush_yds_rank'] = total_rush_with_rank_19['rush_yds'].rank(ascending=False, method='min')
rush_success_with_rank_19 = all_teams_rush_success_2019.copy()
rush_success_with_rank_19['rush_success_rank'] = rush_success_with_rank_19['rush_success'].rank(ascending=False, method='min')
rush_avg_with_rank_19 = all_teams_rush_avg_2019.copy()
rush_avg_with_rank_19['rush_avg_rank'] = rush_avg_with_rank_19['rush_avg'].rank(ascending=False, method='min')
rush_tds_with_rank_19 = all_teams_rush_tds_2019.copy()
rush_tds_with_rank_19['rush_tds_rank'] = rush_tds_with_rank_19['rush_tds'].rank(ascending=False, method='min')
rush_fumbles_lost_with_rank_19 = all_teams_rush_fumbles_lost_2019.copy()
rush_fumbles_lost_with_rank_19['fumbles_lost_rank'] = rush_fumbles_lost_with_rank_19['fumbles_lost'].rank(ascending=False, method='min')
```

```python
rush_rank_19 = pd.merge(total_rush_with_rank_19[['name', 'rush_yds', 'rush_yds_rank']],
                          rush_avg_with_rank_19[['name', 'rush_avg', 'rush_avg_rank']],
                          on='name')
rush_rank_19 = pd.merge(rush_rank_19,
                          rush_tds_with_rank_19[['name', 'rush_tds', 'rush_tds_rank']],
                          on='name')
rush_rank_19 = pd.merge(rush_rank_19,
                          rush_fumbles_lost_with_rank_19[['name', 'fumbles_lost', 'fumbles_lost_rank']],
                          on='name')
rush_rank_19 = pd.merge(rush_rank_19,
                          rush_success_with_rank_19[['name', 'rush_success', 'rush_success_rank']],
                          on='name')
rush_rank_19.sort_values(by='rush_yds_rank')
```
